```python
import numpy as np
import csv
import math
```

```python
csv_reader = csv.reader(open('view_record.csv', 'r', encoding = 'utf-8'))
```

```python
entertainment_view_ratio = np.zeros(101)
```

```python
print(entertainment_view_ratio)
```

```python
count = 0

for line in csv_reader:
    print(line)
    
    count += 1
    
    if count == 100:
        break
```

```python
for line in csv_reader:
    if line[16] == "예능":
        view_ratio = float(line[15])*100
        view_ratio = math.floor(view_ratio)
    
        entertainment_view_ratio[view_ratio]+=1
```

```python
entertainment_vr_cumsum = np.cumsum(entertainment_view_ratio)

vr_cdf = [(count/entertainment_vr_cumsum[-1]) for count in entertainment_vr_cumsum]
```

```python
fig, ax = plt.subplots()
xrange = 101
plt.plot(vr_cdf[:xrange], 'r-', label='entertainment')

plt.legend(loc=4)
plt.xlabel('time per seek(m)')
plt.ylabel('CDF')
plt.xlim(0, xrange)
ax.set_xticks(np.arange(0, 101, 10))
plt.grid()
plt.show()
```
