```python
import csv
import pandas as pd
import numpy as np
```

```python
columnName = ['uno', 'movie2', 'movie2_viewtime', 'movie1', 'movie1_viewtime']
viewtime_for_compare = pd.read_csv('viewtime_for_compare.csv', names = columnName)
```

```python
distinct_movie_2 = np.unique(viewtime_for_compare['movie2'])
```

```python
distinct_movie_1 = np.unique(viewtime_for_compare['movie1'])
```

```python
print(distinct_movie_1)
```

```python
# for each movie 2

w_matrix_columns = ['movie_1', 'movie_2', 'weight']
w_matrix=pd.DataFrame(columns=w_matrix_columns)

for movie_1 in distinct_movie_1:
    mv1_tmpset = viewtime_for_compare[viewtime_for_compare['movie1'] == movie_1]
    distinct_movie_2 = np.unique(mv1_tmpset['movie2'])
    for movie_2 in distinct_movie_2:
        paired_movie_1_2 = mv1_tmpset[mv1_tmpset['movie2'] == movie_2]
        sim_value_numerator = (paired_movie_1_2['movie1_viewtime'] * paired_movie_1_2['movie2_viewtime']).sum()
        sim_value_denominator = np.sqrt(np.square(paired_movie_1_2['movie1_viewtime']).sum()) 
                                * np.sqrt(np.square(paired_movie_1_2['movie2_viewtime']).sum())
        sim_value_denominator = sim_value_denominator if sim_value_denominator != 0 else 1e-8
        sim_value = sim_value_numerator / sim_value_denominator
        w_matrix = w_matrix.append(pd.Series([movie_1, movie_2, sim_value], index=w_matrix_columns), 
                                   ignore_index=True)
```

```python
w_matrix.head(100)
```

```python

# calculate the predicted ratings
def predict(userId, movieId):
    if avg_viewtime[avg_viewtime['contentId'] == movieId].shape[0] > 0:
        mean_rating = avg_viewtime[avg_viewtime['contentId'] == movieId]['avg_viewtime'].iloc[0]
    else:
        mean_rating = 2.5
    
    
    user_other_ratings = viewtime_for_compare[viewtime_for_compare['uno'] == userId]
    user_distinct_movies = np.unique(user_other_ratings['movie2'])
    sum_weighted_other_ratings = 0
    sum_weghts = 0
    
    for movie_j in user_distinct_movies:
        if avg_viewtime[avg_viewtime['contentId'] == movie_j].shape[0] > 0:
            rating_mean_j = avg_viewtime[avg_viewtime['contentId'] == movie_j]['avg_viewtime'].iloc[0]
        else:
            rating_mean_j = 2.5
       # only calculate the weighted values when the weight between movie_1 and movie_2 exists in weight matrix
    
        w_movie_1_2 = w_matrix[(w_matrix['movie_1'] == movieId) & (w_matrix['movie_2'] == movie_j)]
        if w_movie_1_2.shape[0] > 0:
            user_rating_j = user_other_ratings[user_other_ratings['movieId']==movie_j]
            sum_weighted_other_ratings += (user_rating_j['movie2_viewtime'].iloc[0] - rating_mean_j) 
            * w_movie_1_2['weight'].iloc[0]
            sum_weghts += np.abs(w_movie_1_2['weight'].iloc[0])
        
    # if sum_weights is 0 (which may be because of no ratings from new users), use the mean ratings
    if sum_weghts == 0:
        predicted_rating = mean_rating
        # sum_weights is bigger than 0
    else:
        predicted_rating = mean_rating + sum_weighted_other_ratings/sum_weghts
        
    return predicted_rating
```
