```python
import csv
```

```python
csv_reader = csv.reader(open('view_record/daily_view_record_on_07.csv', 'r', encoding = 'utf-8'))
```

```python
count = 0

for line in csv_reader:
    print(line)
    
    count += 1
    
    if count == 100:
        break
```

```python
prev_line = {}
datalist_by_uno = []

f = open('view_record_on_07.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(f)

for line in csv_reader:
    # read current line
    line_dictionary = {'date':line[0], 'hour':int(line[1]), 'uno':line[2], 'contentId':line[3], 
                       'devicetype':line[4],'ipaddress:':line[5], 'viewtime':int(line[6]), 
                       'start_runtime': int(line[7]), 'stop_runtime':line[8],'start_timestamp':int(line[9]),
                       'stop_timestamp':int(line[10]), 'seek':int(line[11]),'pause':int(line[12]), 'stop':int(line[13])}
    if prev_line == {}:
        prev_line = line_dictionary
        datalist_by_uno.append(line)
        continue
        
    last_record = datalist_by_uno[-1:][0]
    # compare previous uno and current uno
    if prev_line['uno'] == line_dictionary['uno']:
        # if two uno is equal
        # collect rows of this uno
        
        if prev_line['contentId'] == line_dictionary['contentId']:
            # check that prev row and current row are continuous
            if (int(line_dictionary['hour']) - int(prev_line['hour']) == 1) and (int(line_dictionary['start_timestamp']) - int(prev_line['stop_timestamp']) < 60): 
                #merge two rows
                new_line = line
                new_line[1] = last_record[1] # set hour
                new_line[6] = str(int(last_record[6])+int(line_dictionary['viewtime'])) # set viewtime
                new_line[7] = last_record[7] #set start_runtime
                new_line[9] = last_record[9] #set start_timestamp
                new_line[11] = str(int(last_record[11])+int(line_dictionary['seek']))
                new_line[12] = str(int(last_record[12])+int(line_dictionary['pause']))
                new_line[13] = str(int(last_record[13])+int(line_dictionary['stop']))
                
                datalist_by_uno[len(datalist_by_uno)-1] = new_line
        else:
            datalist_by_uno.append(line)
    else:
        for row in datalist_by_uno:
            wr.writerow(row)

        datalist_by_uno = list()
        datalist_by_uno.append(line)
            
    prev_line = line_dictionary
    
f.close()
```
